# SGD

```python
with open('edmd_sgd.pkl', 'rb') as inp:
    edmd_sgd = pickle.load(inp)
```

```python
plt.plot(np.abs(edmd_sgd.koopman_eigenvalues), 'x')
plt.title('Eigenvalues SGD');
plt.savefig("sgd.png")
```

```python
sum(np.abs(edmd_sgd.koopman_eigenvalues) > 0.05)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax11 = fig.add_subplot(2, 3, 1)   #top left
ax22 = fig.add_subplot(2, 3, 2 )
ax33 = fig.add_subplot(2, 3, 3 )

ax2 = fig.add_subplot(2, 2, 3)   #top right
ax3 = fig.add_subplot(2, 2, 4)   #bottom left

ax11.plot(np.sort(np.abs(edmd_adam.koopman_eigenvalues)), 'x', alpha=0.2,c = "black")
ax22.plot(np.sort(np.abs(edmd_sgd_mom_05.koopman_eigenvalues)), 'x',alpha=0.2, c = "black")
ax33.plot(np.sort(np.abs(edmd_sgd.koopman_eigenvalues)), 'x', alpha=0.1, label="SGD", c = "black")

ax11.set_ylabel("eigen value(absolute)")
ax11.title.set_text("ADAM")
ax11.title.set_fontsize(9)
ax22.title.set_text("SGD-momentum")
ax22.title.set_fontsize(9)
ax33.title.set_text("SGD")
ax33.title.set_fontsize(9)




ax2.hist(np.abs(edmd_adam.koopman_eigenvalues), bins=100, label="ADAM")
ax2.hist(np.abs(edmd_sgd_mom_05.koopman_eigenvalues), bins=100, label="SGD-momentum")
ax2.hist(np.abs(edmd_sgd.koopman_eigenvalues), bins=100, label="SGD")

edmd_adam_eigs_subs = np.abs(edmd_adam.koopman_eigenvalues)[np.abs(edmd_adam.koopman_eigenvalues)>0.1]
edmd_sgd_eigs_subs = np.abs(edmd_sgd.koopman_eigenvalues)[np.abs(edmd_sgd.koopman_eigenvalues)>0.1]
edmd_sgd_momentum_subs = np.abs(edmd_sgd_mom_05.koopman_eigenvalues)[np.abs(edmd_sgd_mom_05.koopman_eigenvalues)>0.1]

ax3.hist(edmd_adam_eigs_subs, bins=100, label="ADAM")
ax3.hist(edmd_sgd_eigs_subs, bins=100, label="SGD", color="#2ca02c")
# plt.hist(edmd_sgd_momentum_subs, bins=100, label="SGD-momentum")


ax3.set_xlabel("eigen value(absolute)")
ax3.legend(loc="upper left")

ax2.set_xlabel("eigen value(absolute)")
ax2.set_ylabel("count")
ax2.legend(loc="upper right")

plt.show()
fig.savefig("eigenvalue_plots.png", bbox_inches='tight')

```
